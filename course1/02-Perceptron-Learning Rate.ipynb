```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def h(X):
    a = np.dot(X,w)+b
    return (a>=0).astype(int)

w = np.array([0,0])
b = 0.1
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
z = np.array([0,0,0,1])
```

```python
plt.gca(aspect=1)
plt.scatter(X[:,0],X[:,1],100,c=z,edgecolor='r',marker='D',cmap='hot')
plt.show()
```

หา w,b ที่ทำให้ h(X)==  ทุกตัว

** η (learning rate)

```python
eta = 0.2 # อัตราการเรียนรู้
print('เริ่มต้น: h(x)=%s, w=%s, b=%s'%(h(X),w,b))
for j in range(100): # ให้ทำซ้ำสูงสุด 100 ครั้ง
    for i in range(4):
        z_h = z[i] - h(X[i])
        dw = eta*z_h*X[i]
        db = eta*z_h
        #w += dw
        np.add(w,dw,out=w,casting="unsafe")
        b += db
        print('รอบ %d.%d: h(x)=%s, w=%s, b=%s, Δw=%s, Δb=%s'%(j+1,i+1,h(X),w,b,dw,db))
    if(np.all(h(X)==z)): # ถ้าผลลัพธ์ถูกต้องทั้งหมดก็ให้เสร็จสิ้นการเรียนรู้
        break
```

Typically this is code such as a += b and you need to make it np.add(a, b, out=a, casting="unsafe") (if you are sure you want the unsafe cast behaviour. Or, unless b is huge and it is safe, maybe just cast b first.
